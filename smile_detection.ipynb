# Importing the necessary libraries :

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Now reading the image & view it :

```python
yassir=cv2.imread('yassir_smile.png',0)
```

```python
plt.imshow(yassir,cmap='gray')
```

# Now using the needed cascade :

```python
smile_cascade=cv2.CascadeClassifier('haarcascades/haarcascade_smile.xml')
```

# Now let's create the function to detect the smile :

```python
def smile_face(img):
    smile_img=img.copy()
    smile_rects=smile_cascade.detectMultiScale(smile_img)
    for (x,y,w,h) in smile_rects:
        cv2.rectangle(smile_img,(x,y),(x+w,y+h),(255,255,255),10)
    return smile_img
```

# Now let's apply this function on that image :

```python
result=smile_face(yassir)
```

# Now its the time to view the image :

```python
plt.imshow(result,cmap='gray')
```

# The result is that so many things has been detected as a smile...In this case we need to add some parameters to the smile_face function such as scaleFactor & minNeighbors on the detectMultiScale command :

```python
def smile_face_2(img):
    smile_img=img.copy()
    smile_rects=smile_cascade.detectMultiScale(smile_img,scaleFactor=10,minNeighbors=8)
    for (x,y,w,h) in smile_rects:
        cv2.rectangle(smile_img,(x,y),(x+w,y+h),(255,255,255),10)
    return smile_img
```

# Now let's apply the new function to the image :

```python
result=smile_face_2(yassir)
```

# Let's view the result :

```python
plt.imshow(result,cmap='gray')
```
