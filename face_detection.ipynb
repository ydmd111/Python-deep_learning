# Importing the necessary libraries :

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Now is to read the image and view it in gray scale :

```python
yassir=cv2.imread('yassir2.jpg',0)
```

```python
plt.imshow(yassir,cmap='gray')
```

# Uploading the cascade for detecting the face :

```python
face_cascade=cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_default.xml')
```

# Now let's create a function to detect the face with tringle :

```python
def detect_face(img):
    face_img=img.copy()
    face_rects=face_cascade.detectMultiScale(face_img)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img
```

# Let's apply the function on the yassir's image :

```python
result=detect_face(yassir)
```

# Let's view the detected face's image :

```python
plt.imshow(result,cmap='gray')
```
