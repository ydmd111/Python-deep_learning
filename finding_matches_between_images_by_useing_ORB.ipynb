# Importing the necessary libraries :

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Creating a function to display images :

```python
def display_img(img,cmap='gray'):
    fig=plt.figure(figsize=(10,12))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

# Reading the first image and display it :

```python
reeses=cv2.imread('reeses_puffs.png',0)
```

```python
display_img(reeses)
```

# Now is the second image :

```python
cereals=cv2.imread('many_cereals.jpg',0)
```

```python
display_img(cereals)
```

# Now creating the detector object :

```python
orb=cv2.ORB_create()
```

# Now we find the key points & descriptors of this object :

```python
kp1,des1=orb.detectAndCompute(reeses,None)
```

```python
kp2,des2=orb.detectAndCompute(cereals,None)
```

# Creating the matching object by brute-force method :

```python
bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
```

# Now is to check where the matching occurs & display them :

```python
matches=bf.match(des1,des2)
```

```python
matches
```

# Now let's view the first matching & view it's distance :

```python
single_match=matches[0]
```

# When the distance value is less, means the the matching is better :

```python
single_match.distance
```

# Let's sort the matches based on the distance :

```python
matches=sorted(matches,key=lambda x:x.distance)
```

# Let's view the length of the matches :

```python
len(matches)
```

# Now let's draw the matches of the first 25 only :

```python
reeses_matches=cv2.drawMatches(reeses,kp1,cereals,kp2,matches[:25],None,flags=2)
```

# Now let's display the draw matches :

```python
display_img(reeses_matches)
```
