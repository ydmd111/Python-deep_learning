# Importing the cv2 library :

```python
import cv2
```

# Capturing video via the laptop camera in gray style :

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame',gray)
    if cv2.waitKey(1)& 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# Capturing video via the laptop camera in colored style :

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    cv2.imshow('frame',frame)
    if cv2.waitKey(1)& 0xFF==ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

# To view the screenshots for the outputs..First import matplotlib.pyplot library :

```python
import matplotlib.pyplot as plt
```

# The first output (the gray output) :

```python
gray_img=cv2.imread('gray.png')
gray_img=cv2.cvtColor(gray_img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img,cmap='gray')
```

# The second output(the colored one)

```python
colored_img=cv2.imread('colored.png')
colored_img=cv2.cvtColor(colored_img,cv2.COLOR_BGR2RGB)
plt.imshow(colored_img)
```
